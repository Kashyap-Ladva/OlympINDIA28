```python
import pandas as pd

medallists = pd.read_csv("../data/raw/medallists.csv")

medallists.columns
```

```python
paris = medallists[medallists["medal_type"].notna()].copy()
```

```python
paris_final = paris.rename(columns={
    "name": "Name",
    "gender": "Sex",
    "country_code": "NOC",
    "country": "Country",
    "discipline": "Sport",
    "event": "Event",
    "medal_type": "Medal"
})

paris_final["Year"] = 2024
```

```python
paris_final = paris_final[
    ["Name", "Sex", "NOC", "Country", "Sport", "Event", "Medal", "Year"]
]
```
