## Phase 1: Data Loading & Validation

- Loaded Olympic athlete-event data (1896–2016)
- Validated schema, medal distribution, and year coverage
- Mapped NOC codes to country names
- Confirmed India-specific data presence
- Saved cleaned base dataset for further analysis


```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

####Load DataSets


```python
# Core Olympic dataset
olympic = pd.read_csv("../data/raw/olympic_history.csv")

# Lookup table
countries = pd.read_csv("../data/raw/country_codes.csv")

# Aggregated medal data
medal_tally = pd.read_csv("../data/raw/olympic_games_medal_tally.csv")

# External indicators
population = pd.read_csv("../data/raw/population.csv")
socio = pd.read_csv("../data/raw/socioeconomics.csv")

#Shapes of All DataSets
print("Olympic history:", olympic.shape)
print("Country codes:", countries.shape)
print("Medal tally:", medal_tally.shape)
print("Population:", population.shape)
print("Socioeconomic:", socio.shape)

```

```python
olympic = olympic.merge(
    countries,
    left_on="NOC",
    right_on="NOC",
    how="left"
)
```

```python
olympic[olympic["NOC"] == "IND"].head()
```

```python
base_cols = [
    "ID","Name","Sex","Age","Height","Weight",
    "Year","Season","Sport","Event",
    "Medal","NOC","region"
]

base_df = olympic[base_cols]

base_df.to_csv("../data/processed/olympic_base.csv", index=False)

print("Saved olympic_base.csv:", base_df.shape)

```
