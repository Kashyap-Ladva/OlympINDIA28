```python
import pandas as pd

tokyo = pd.read_csv("../data/raw/tokyo_2020_final.csv")
paris = pd.read_csv("../data/raw/paris_2024_final.csv")

print("TOKYO 2020")
print(tokyo.shape)
print(tokyo.columns)
print(tokyo["Medal"].value_counts(dropna=False))

print("\nPARIS 2024")
print(paris.shape)
print(paris.columns)
print(paris["Medal"].value_counts(dropna=False))
```

```python
def normalize_medal(df):
    df["Medal"] = df["Medal"].str.replace(" Medal", "", regex=False)
    return df

tokyo = normalize_medal(tokyo)
paris = normalize_medal(paris)

print(tokyo["Medal"].value_counts())
print(paris["Medal"].value_counts())
```

```python
cols = ["Name", "Sex", "NOC", "Year", "Sport", "Event", "Medal"]
tokyo = tokyo[cols]
paris = paris[cols]
```

```python
history = pd.read_csv("../data/raw/olympic_history.csv")

history = history[
    ["Name", "Sex", "NOC", "Year", "Sport", "Event", "Medal"]
]
```

```python
full_olympics = pd.concat(
    [history, tokyo, paris],
    ignore_index=True
)

print(full_olympics.shape)
print(full_olympics["Year"].value_counts().sort_index())
```

```python
full_olympics["Medal"].value_counts(dropna=False)
```

```python
full_olympics.to_csv(
    "../data/processed/olympic_1896_2024_extended.csv",
    index=False
)
```
